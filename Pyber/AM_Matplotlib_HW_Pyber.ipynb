```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
df_city = pd.read_csv(city_data_to_load)
df_ride = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
df = pd.merge(df_ride, df_city, how ='left', on='city')

#Display the data table for preview
df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
numberofrides = df.groupby("city")["ride_id"].count()
avgfare = df.groupby("city")["fare"].mean()
drivercount = df.groupby("city")["driver_count"].mean()  

# Build the scatter plots for each city types
#Urban = plt.scatter(numberofrides, avgfare, s=[drivercount*5],c= 'r', edgecolors="black", alpha =.4, label ="Urban")
#Suburban = plt.scatter(numberofrides, avgfare, s=[drivercount*5],c= 'b', edgecolors="black", alpha =.4, label ="Subrban")
Rural = plt.scatter(numberofrides, avgfare, s=[drivercount*5], c='g', edgecolors="black", alpha =.4, label = "Rural")

# Incorporate the other graph properties
plt.title("Ride share Data (2016)", fontsize=16)
plt.xlabel("Number of Rides (per city)", fontsize=12)
plt.ylabel("Average Fare ($)", fontsize=12)
plt.grid()

# Create a legend
#plt.legend(handles =['Urban', 'Suburban', 'Rural'], loc="best")

# Incorporate a text label regarding circle size


# Save Figure
plt.savefig('rides.png')
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
farebylocation = df.groupby("type")["fare"].sum()
labels = ["Rural", "Suburban","Urban"]
colors = ["yellow", "lightskyblue","lightcoral"]
explode = (0,0, 0.1)

# Build Pie Chart
plt.pie(farebylocation,explode = explode, labels=labels, colors=colors, autopct="%2.1f%%", startangle=140)

# Save Figure
plt.savefig('fares.png')

```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
ridesbylocation = df.groupby("type")["ride_id"].count()
labels = ["Rural", "Suburban","Urban"]
colors = ["yellow", "lightskyblue","lightcoral"]
explode = (0,0, 0.1)

# Build Pie Chart
plt.pie(ridesbylocation,explode = explode, labels=labels, colors=colors, autopct="%2.1f%%", startangle=140)

# Save Figure
plt.savefig('ridespie.png')

```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
driverbylocation = df.groupby("type")["driver_count"].sum()
labels = ["Rural", "Suburban","Urban"]
colors = ["yellow", "lightskyblue","lightcoral"]
explode = (0,0, 0.1)

# Build Pie Chart
plt.pie(driverbylocation,explode = explode, labels=labels, colors=colors, autopct="%2.1f%%", startangle=140)

# Save Figure
plt.savefig('drivers.png')
```

```python
# Show Figure
plt.show()
```
